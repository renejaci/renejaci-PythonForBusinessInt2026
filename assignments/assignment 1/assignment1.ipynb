## Task 1: A list of dictionaries

```python
# Given transaction data
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

# a) Filter out invalid transactions (price is zero or quantity is negative)
valid_transactions = []
for t in transactions:
    if t["price"] > 0 and t["quantity"] > 0:
        valid_transactions.append(t)

# b) Compute a new field 'total' for each valid transaction
for t in valid_transactions:
    t["total"] = t["price"] * t["quantity"]

# c) Group by region and compute total revenue per region
revenue_by_region = {}
for t in valid_transactions:
    region = t["region"]
    if region in revenue_by_region:
        revenue_by_region[region] += t["total"]
    else:
        revenue_by_region[region] = t["total"]

print("Revenue by region:", revenue_by_region)

# d) Find and print the region with the highest total revenue
best_region = ""
best_revenue = 0
for region, rev in revenue_by_region.items():
    if rev > best_revenue:
        best_revenue = rev
        best_region = region

print(f"The region with the highest revenue is {best_region} with {best_revenue}€")
```

## Task 2: Creating a function

```python
# Given sales data
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(daily_sales):
    """Takes a list of daily sales values and returns a summary dictionary."""
    
    total = sum(daily_sales)
    average = round(total / len(daily_sales), 2)
    
    # best_day and worst_day use index starting from 1
    best_day = daily_sales.index(max(daily_sales)) + 1
    worst_day = daily_sales.index(min(daily_sales)) + 1
    
    # Count days where sales exceeded the average
    above_average_count = 0
    for sale in daily_sales:
        if sale > average:
            above_average_count += 1
    
    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    }

# Call the function and print the result
result = summarise_sales(sales)
print(result)
```

## Task 3: Loops

```python
# Given student records
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

# a) Create a list of students with grade 50 or above
passing = []
for student in students:
    if student["grade"] >= 50:
        passing.append(student["name"])

print("Passing students:", passing)

# b) Print a status message for each student
for student in students:
    if student["grade"] >= 80:
        print(f"{student['name']}: Excellent")
    elif student["grade"] >= 50:
        print(f"{student['name']}: Pass")
    else:
        print(f"{student['name']}: Fail")

# c) Count students with attendance below 0.80
low_attendance_count = 0
for student in students:
    if student["attendance"] < 0.80:
        low_attendance_count += 1

print(f"\nStudents with attendance below 0.80: {low_attendance_count}")

# d) Find the student with the highest grade using a loop (no max())
top_student = students[0]
for student in students:
    if student["grade"] > top_student["grade"]:
        top_student = student

print(f"Highest grade (loop): {top_student['name']} with grade {top_student['grade']}")

# e) Find the student with the highest grade using max()
top_student_max = max(students, key=lambda s: s["grade"])
print(f"Highest grade (max): {top_student_max['name']} with grade {top_student_max['grade']}")
```

## Task 4: Class and methods

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

    # a) Method to add or update an asset
    def add_asset(self, ticker, shares, price_per_share):
        """Adds a new asset or updates shares and price if ticker already exists."""
        self.holdings[ticker] = {
            "shares": shares,
            "price_per_share": price_per_share
        }

    # b) Method to calculate total portfolio value
    def total_value(self):
        """Returns the total market value of all holdings."""
        total = 0
        for ticker in self.holdings:
            asset = self.holdings[ticker]
            total += asset["shares"] * asset["price_per_share"]
        return total

    # c) Method to find the most valuable asset
    def most_valuable(self):
        """Returns the ticker of the asset with the highest total value."""
        best_ticker = ""
        best_value = 0
        for ticker in self.holdings:
            asset = self.holdings[ticker]
            value = asset["shares"] * asset["price_per_share"]
            if value > best_value:
                best_value = value
                best_ticker = ticker
        return best_ticker

    # d) String representation of the portfolio
    def __str__(self):
        return f"Portfolio of {self.owner}: {len(self.holdings)} assets, total value: {self.total_value()}€"

# Create an instance and demonstrate all methods
my_portfolio = Portfolio("Alice")

my_portfolio.add_asset("AAPL", 10, 180)
my_portfolio.add_asset("MSFT", 5, 420)
my_portfolio.add_asset("TSLA", 8, 250)

print(my_portfolio)  # Uses the __str__ method
print(f"Total value: {my_portfolio.total_value()}€")
print(f"Most valuable asset: {my_portfolio.most_valuable()}")

# Demonstrate updating an existing asset
my_portfolio.add_asset("AAPL", 15, 185)
print(f"\nAfter updating AAPL:")
print(my_portfolio)
```

## Task 5: Importing a module

```python
from datetime import datetime

# Given order data
orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

# Parse each date string into a datetime.date object
dates = []
for order in orders:
    date = datetime.strptime(order[0], "%Y-%m-%d").date()
    dates.append(date)

# Find the earliest and latest order dates
earliest = min(dates)
latest = max(dates)

# Compute and print the number of days between them
difference = latest - earliest
print(f"Earliest order: {earliest}")
print(f"Latest order: {latest}")
print(f"Days between earliest and latest order: {difference.days}")
```

## Task 6: Array operations and broadcasting

```python
import numpy as np

# Quarterly revenue data (rows: years 2022-2024, columns: Q1-Q4)
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]])

# a) Total annual revenue for each year (row sums)
annual_revenue = revenue.sum(axis=1)
print("Annual revenue per year:", annual_revenue)

# b) Mean quarterly revenue across all years (column means)
quarterly_means = revenue.mean(axis=0)
print("Mean quarterly revenue:", quarterly_means)

# c) Apply growth adjustment to 2024 row using broadcasting
growth_factors = np.array([1.05, 1.03, 1.08, 1.02])
adjusted_2024 = revenue[2] * growth_factors
print("Adjusted 2024 revenue:", adjusted_2024)

# d) Boolean indexing: find values exceeding 250
above_250 = revenue[revenue > 250]
print("Quarterly revenues exceeding 250:", above_250)
```

## Task 7: Pandas

```python
import pandas as pd

# Create the DataFrame
data = {
    "employee": ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department": ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary": [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9]
}

df = pd.DataFrame(data)

# a) Add salary_per_year_exp column, rounded to nearest integer
df["salary_per_year_exp"] = (df["salary"] / df["years_exp"]).round(0).astype(int)
print("DataFrame with salary per year of experience:")
print(df)

# b) Sort by performance_score descending and print top 3
sorted_df = df.sort_values("performance_score", ascending=False)
print("\nTop 3 employees by performance score:")
print(sorted_df.head(3))

# c) Using .loc, select Finance employees with performance score above 4.0
finance_high = df.loc[(df["department"] == "Finance") & (df["performance_score"] > 4.0)]
print("\nFinance employees with performance score above 4.0:")
print(finance_high)

# d) Rename years_exp to experience_years
df = df.rename(columns={"years_exp": "experience_years"})
print("\nColumns after rename:")
print(df.columns.tolist())

# e) Using .iloc, select salary and performance_score for first four employees
print("\nSalary and performance score for first 4 employees:")
print(df.iloc[:4, [2, 4]])
```

## Task 8: More Pandas

```python
import numpy as np
import pandas as pd

# Generate the data as instructed
rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

# a) Create the DataFrame with date as index
traffic = pd.DataFrame({
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
}, index=days)

traffic.index.name = "date"

print("Website traffic DataFrame:")
print(traffic.head())

# b) Compute new columns: conversions and daily_revenue
traffic["conversions"] = (traffic["visitors"] * traffic["conversion_rate"]).round(0).astype(int)
traffic["daily_revenue"] = traffic["conversions"] * traffic["avg_order_value"]

print("\nDataFrame with conversions and daily revenue:")
print(traffic.head(10))
```
